```python
import numpy as np
from scipy import stats
```

```python
# Define monthly sales data 
sales_data = np.array([15000, 18000, 17000, 16000, 20000, 19000, 17000, 2100])
```

```python
# Compute mean, median, and mode 
mean_sales = np.mean(sales_data) 
median_sales = np.median(sales_data) 
mode_sales = stats.mode (sales_data, keepdims=True).mode[0]
```
